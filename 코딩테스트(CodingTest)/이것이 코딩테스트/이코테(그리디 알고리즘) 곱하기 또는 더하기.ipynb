# 그리디 알고리즘 곱하기or 더하기

```python
## 12/13 복습 (완)

data = input()

result = int(data[0])


for i in range(1,len(data)):
    
    num = int(data[i])
    
    if num<=1 or result<=1:
        result+=num
    else:
        result*=num
        
print(result)
```

# 공포도 알고리즘


이 알고리즘은 "공포도"를 기준으로 모험가 그룹을 구성하는 것으로 보입니다. 

여기서 각 모험가의 공포도는 그 모험가가 속한 그룹의 최소 인원수를 의미합니다. 

알고리즘은 다음과 같이 작동합니다:

1.모험가의 공포도를 나타내는 정수 리스트를 입력받습니다.

2.이 리스트를 오름차순으로 정렬합니다.

3.group (형성된 그룹의 수)과 member (현재 그룹에 속한 모험가의 수) 변수를 0으로 초기화합니다.

4.정렬된 리스트를 순회하며, 각 모험가의 공포도를 확인합니다:

5.각 단계에서 member를 1 증가시킵니다 (새로운 모험가를 현재 그룹에 추가).

6.만약 member의 값이 현재 모험가의 공포도보다 크거나 같다면, 그룹 하나를 형성합니다 (group 증가).

7.이후 member를 0으로 초기화하여 새 그룹을 시작합니다.

8.모든 모험가를 확인한 후, 형성된 총 그룹 수 (group)를 출력합니다.

```python
data = list(map(int,input().split()))

data.sort()

member = 0

group = 0


for i in data:
    member+=1
    
    if member>=i:
        group+=1
        member=0
        
print(group) # 전체 그룹갯수 


print(member) # 현재 그룹에 들어가지 못하고 남아있는 멤버들...
```

공포도 (i): 
알고리즘에서 순회하는 각 모험가의 공포도입니다. 
이 값은 각 단계에서 리스트 data의 현재 요소값을 나타냅니다.



현재 그룹원 수 (member): 
현재 형성 중인 그룹에 포함된 모험가의 수입니다. 
각 단계에서 모험가 한 명이 추가될 때마다 member 값이 1씩 증가합니다.



상태: 
이 컬럼은 알고리즘의 각 단계에서 그룹이 형성되었는지 여부를 나타냅니다. 
만약 member 값이 현재 공포도 (i) 값보다 크거나 같아지면, 그룹이 형성되고 group이 1 증가합니다. 
이 때 '그룹 형성'이라고 표시됩니다. 그렇지 않으면 '-'로 표시됩니다.



그룹 수 (group): 
이 값은 표의 마지막 행에 '총 그룹 수'로 표시됩니다. 
알고리즘의 각 단계에서 새로운 그룹이 형성될 때마다 group 값이 1씩 증가합니다. 
최종적으로 형성된 총 그룹의 수를 나타냅니다.



따라서, 
각 단계에서 i (공포도)는 현재 모험가의 공포도를, 
member는 그 단계까지의 그룹원 수를, 
그리고 group은 전체적으로 형성된 그룹의 수를 나타냅니다.
