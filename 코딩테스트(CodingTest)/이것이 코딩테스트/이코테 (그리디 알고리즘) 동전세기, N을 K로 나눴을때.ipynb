# 잔돈 세기

```python
n= int(input())
count = 0

# 큰단위 화폐부터 차례대로 확인하기 

array=[500,100,50,10]

for coin in array:
    count += (n // coin)   # // 은 나머지가 생략되어짐

    
# count += n // coin: 현재 동전 coin으로 거슬러 줄 수 있는 최대 개수를 구하고,
#이를 count에 더합니다. 예를 들어, 처음에 n이 1260원이고 coin이 500원일 때,
#500원 동전은 2개 필요하므로 count는 2가 됩니다.
    n%= coin
    
#n %= coin: 사용한 동전의 가치를 n에서 빼줍니다. 
#이는 n을 coin으로 나눈 나머지를 n에 할당하는 것과 같습니다. 
#예를 들어, 1260원에서 500원 동전 2개를 사용하고 나면, 
#남은 금액은 260원이 됩니다.

print(count)



```

# N을 K 로 나눴을때 1까지 가는 경로 (큰수찾기)


```python
## 어떠한 수 N이 1이 될때까지 두 과정중 하나를 반복적으로 수행하려고한다.
## 단 , 두번째 연산은 N이 K로 나누어 떨어질때만 선택할수있다.
# 최소 횟수 구하기...
# 나눌수있을때는 나눠야함, 그리고 못나눌때는 어쩔수없이 뺌

n = int(input())
k = 5
result = 0


while n >= k:
    # 한 번에 K로 나누어 떨어지는 가장 가까운 수까지 빼기
    result += (n % k)   # result = 0      
    n -= (n % k)        # n =    20          
    
    if n >= k:
    # 나누기 연산 수행   
        result += 1          #result = 1     
        n //= k              # n = n//k =4        

# 마지막으로 남은 수 만큼 빼기 (N이 1이 될 때까지)
result += (n - 1)              
print(result)
```

```python
#### n = 26 # n을 7로 초기화합니다. 이는 최종적으로 1로 만들어야 하는 초기 값입니다.
k = 3  # k를 3으로 초기화합니다. 이는 나누는 수입니다.
result = 0  # 결과를 저장할 변수를 0으로 초기화합니다. 이는 n을 1로 만드는 데 필요한 최소한의 단계 수를 나타냅니다.

while n >= k:  # n이 k 이상인 경우, 즉 n을 k로 나눌 수 있는 경우 반복합니다.
    
    result += (n % k)  # n을 k로 나눈 나머지를 result에 더합니다. 이는 n을 k의 배수로 만드는 데 필요한 단계 수를 나타냅니다.
    n -= (n % k)  # n에서 나머지를 빼서 k의 배수로 만듭니다.
    
    if n >= k:  # 만약 n이 아직 k 이상이라면
        result += 1  # result에 1을 더합니다. 이는 n을 k로 나누는 단계를 나타냅니다.
        n //= k  # n을 k로 나누어 업데이트합니다.
        
result += (n - 1)  # 마지막으로, n이 1이 될 때까지 필요한 단계 수를 더합니다.
print(result)  # 최종 결과를 출력합니다. 이는 n을 1로 만드는 데 필요한 최소한의 단계 수입니다.
```
