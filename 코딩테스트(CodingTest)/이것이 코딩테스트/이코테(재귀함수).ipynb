```python
# 반복적 으로 만든 재귀함수
def factory_fuc(n):
    result =1
    
    for i in range(1,n+1):
        result*=i
        
    return result

factory_fuc(5)

    
```

```python
# 재귀적으로 만든 함수
def factory_fun2(n):
    if n<=1:
        return 1
    else:
        return n*factory_fun2(n-1)
    
factory_fun2(5)
```
