```python
#      (2)
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]  
]

#      1  
visited = [False]*9


# deque 
from collections import deque


#1.     
#2.  true    
#3.    popleft()   
#4.      
#5.   queue   .




def bfs(graph,start,visited):
    
    # queue     (deque).
    queue = deque([start])
    
    #   start queue , 
    # visited[start] True     .
    visited[start] =True
    

    while queue:
        
        # queue   (popleft)   ,   
        v= queue.popleft()  
        print(v,end='')
        
        #      
        for i in graph[v]:
            if not visited[i]:
        #   queue   .
                queue.append(i)
                visited[i] = True


bfs(graph,1,visited)



# 1
# 238      => 1  3 
# 387      => 2   1,7,8  1   3,8,7 
# 8745     => 3   4,5   8,7,4,5  
# 745      => 8   1,7    7,4,5   
# 456      =>  


# 12387456


```
