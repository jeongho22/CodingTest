```python
# 각 노드가 연결된 정보를 표현 (2차원리스트)
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]  
]

# 각 노드가 방문된 정보를 표현 1차원 리스트 
visited = [False]*9


#함수 deque 임포트
from collections import deque


#1. 방문할 노드를 저장한 데크를 만든다
#2. 방문을 true 설정하여 시작노드를 방문하게 만든다
#3. 노드를 하나씩 꺼내 popleft() 하고 노드를 출력한다
#4. 아직 방문하지 않는 원소들은 큐에 삽입하고
#5. 그 노드를 queue에 추가하고 방문했다고 표시합니다.




def bfs(graph,start,visited):
    
    # queue는 현재 방문할 노드들을 저장하는 데크(deque)입니다.
    queue = deque([start])
    
    # 시작 노드 start를 queue에 추가하고, 
    # visited[start]를 True로 설정하여 시작 노드를 방문했다고 표시합니다.
    visited[start] =True
    

    while queue:
        
        # queue에서 노드를 하나씩 꺼내(popleft) 현재 노드로 설정하고, 이 노드를 출력
        v= queue.popleft()  
        print(v,end='')
        
        # 아직 방문하지 않은 인접한 원소들을 큐에삽입
        for i in graph[v]:
            if not visited[i]:
        # 그 노드를 queue에 추가하고 방문했다고 표시합니다.
                queue.append(i)
                visited[i] = True


bfs(graph,1,visited)



# 1
# 238      => 1의 인접노드 3개 
# 387      => 2의 인접 노드는 1,7,8 인데 1은 이미 없어졌으니 3,8,7 이된다
# 8745     => 3의 인접 노드는 4,5 이며 남은노드 8,7,4,5가 된다 
# 745      => 8의 인접 노드는 1,7 인데 이미 처리되서 7,4,5 가 된다 
# 456      => 완료 


# 12387456


```

```python
# 각 노드가 연결된 정보를 표현 (2차원리스트)
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]  
]

# 각 노드가 방문된 정보를 표현 1차원 리스트 
visited = [False]*9


#함수 deque 임포트
from collections import deque


#1. 방문할 노드를 저장한 데크를 만든다
#2. 방문시작을 true 설정하여 시작노드를 방문하게 만든다
#3. while 반복문을 통해 노드를 하나씩 꺼내 popleft() 하고 노드를 출력한다
#4. 그래프안에 있는 노드들을 아직 방문하지 않는 원소들은 
#5. 그 노드를 queue에 추가하고 방문했다고 표시합니다.


def func1(graph,start,visited):
    
    queue = deque([start])
    visited[start] = True
    
    while queue:
        
        v= queue.popleft()
        print(v,end='')
        
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
        
    

func1(graph,1,visited)
```
