```python
n = int(input())   # 1. 공간 N 설정하기 

plans = input().split() # 2. LRUD 인풋 넣기

x,y =2,2            # 3. 최초 좌표 


dx =[0,0,-1,1]      # 위 ,아래
dy =[-1,1,0,0]      # 왼쪽 ,오른쪽 

move =["L","R","U","D"]



# 이동 계획 확인
for plan in plans:
    # 이동 후 좌표 구하기
    for i in range(len(move)):
        if plan == move[i]:
            nx = x +dx[i]
            ny = y +dy[i]
            
    # 공간 벗어나는 경우 무시
    if nx<1 or ny<1 or nx>n or ny> n:
        continue
    # 이동 수행    
    x,y = nx,ny
    
print(x,y)
```

```python
#1. 공간설정
n = int(input())

#2. 좌표 기입

plans = input().split()


#3. 처음 좌표 

x,y =1,1

#4. 이동경로(왼,오,위,아래)


dx=[0,0,-1,1]  # 행       # 위 아래는 행이 움직여야함
dy=[-1,1,0,0]  # 열       # 왼쪽 오른쪽은 열이 움직여야함 




#5.좌표 넣기

move=["L","R","U","D"]

#6.

for plan in plans:
    for i in range(len(move)):
        if plan == move[i]:
            new_x = x+ dx[i]
            new_y = y+ dy[i]
    
        # 공간 벗어나는 경우 무시
    if new_x<1 or new_y<1 or new_x>n or new_y> n:
        continue
            
    # 이동 수행    
    x,y = new_x,new_y
    
print(x,y)
```

5. 이동 경로 dx(x축),dy(y축)

dx =[0,0,-1,1]
dy =[-1,1,0,0]




```python
n = int(input())

plans= input().split()

x,y = 1,1

move =["L","R","U","D"]

dx = [0,0,-1,1]
dy = [-1,1,0,0]


for plan in plans:
    for i in range(len(move)):    ## LRUD 에따라 계속해서 
        if plan == move[i]:       ## 조건에따라 이동함
            new_x = x+ dx[i]
            new_y = y+ dy[i]
        
    if new_x<1 or new_y<1 or new_x>n or new_y>n :
        continue  ## 새로운 좌표가 1보다 작거나 , 범위보다 크면 멈춘다.

    x,y = new_x,new_y

print(x,y)
```
