1. 탐색 시작 노드를 스택에 삽입하고 방문처리한다.
2. 스택의 최상단 노드에 방문하지않은 인접한 노드가 하나라도 있으면 그 노드를 스택에 넣고 방문 처리한다. 방문하지 않는 노드가없으면 최상단 노드를 꺼낸다.
3. 더 이상 2번의 과정을 수행할 수 없을때까지 반복한다.


```python
# 그래프를 준비한다. 방문기준 번호가 낮은 인접노드부터 방문한다


# dfs 메서드 정의
def dfs(graph, v, visited):
    #현재의 노드를 방문처리
    visited[v] = True

    print(v, end='') 
    # 방문했다는 증거로 print 출력을함
    
    #     현재 노드 v에 인접한 노드들을 순회합니다. for i in graph[v]는 v와 연결된 모든 노드를 순회합니다.
    for i in graph[v]: 
        if not visited[i]: ## 인접한 노드가 방문한 상태가 아니면 다시 재귀로 방문함.
            dfs(graph,i,visited)
            


            # 각 노드가 연결된 정보를 표현 2차원리스트로, [] 시작하는부분임
graph=[
    [],      #노드 1
    [2,3,8], #노드 1과 연결되어있는것들
    [1,7],   #노드 2와 연결되어있는것들
    [1,4,5], #노드 3과 연결되어있는것들
    [3,5],   #노드 4와 연결되어있는것들
    [3,4],   #노드 5와 연결되어있는것들 
    [7],     #노드 6과 연결되어있는것들
    [2,6,8], #노드 7과 연결되어있는것들
    [1,7]    #노드 8과 연결되어있는것들
]

# 각 노드가 방문된 정보를 표현 1차원리스트

visited = [False]*9



# 정의된 dfs 함수 호출

dfs(graph,1,visited)
```

# 1. 현재의 노드를 방문처리
# 2. 방문했다는 증거로 print 출력을함
# 3.  각 노드가 방문된 정보를 표현 1차원리스트

```python
# 각 노드가 연결된 정보를 표현 2차원리스트로, [] 시작하는부분임
graph=[
    [],      #노드 1
    [2,3,8], #노드 1과 연결되어있는것들
    [1,7],   #노드 2와 연결되어있는것들
    [1,4,5], #노드 3과 연결되어있는것들
    [3,5],   #노드 4와 연결되어있는것들
    [3,4],   #노드 5와 연결되어있는것들 
    [7],     #노드 6과 연결되어있는것들
    [2,6,8], #노드 7과 연결되어있는것들
    [1,7]    #노드 8과 연결되어있는것들
]

# 각 노드가 방문된 정보를 표현 1차원리스트

visited = [False]*9



def func(graph,v,visited):
    visited[v] =True
    
    print(v, end='') 
    
    # 그래프를 순환한다.
    for i in graph[v]:
        #인접한 노드가 방문상태가 아니라면 다시 재귀로 방문한다.
        if not visited[i]:
            func(graph,i,visited)
            
            
func(graph,1,visited)
```

```python
# 각 노드가 연결된 정보를 표현 2차원리스트로, [] 시작하는부분임
graph=[
    [],      #노드 1
    [2,3,8], #노드 1과 연결되어있는것들
    [1,7],   #노드 2와 연결되어있는것들
    [1,4,5], #노드 3과 연결되어있는것들
    [3,5],   #노드 4와 연결되어있는것들
    [3,4],   #노드 5와 연결되어있는것들 
    [7],     #노드 6과 연결되어있는것들
    [2,6,8], #노드 7과 연결되어있는것들
    [1,7]    #노드 8과 연결되어있는것들
]

# 각 노드가 방문된 정보를 표현 1차원리스트

visited = [False]*9




def func2(graph,v,visited):
    
    visited[v] = True # 방문했을때 true 로바뀜
    print(v,end='')   # 방문한거 출력
    
    for i in graph[v]: # 그래프를 돌면서
        if not visited[i]: # 방문하지않았으면 
            func2(graph,i,visited) # 재귀로 돌아감
            
            
func2(graph,1,visited)
```
