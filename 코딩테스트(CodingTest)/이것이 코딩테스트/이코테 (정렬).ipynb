# 선택정렬

```python
a= [7,5,9,0,3,1,6,2,4,8]

for i in range(len(a)):
    mini = i                  ## 가장 작은 원소의 인덱스
    for j in range(i+1,len(a)):
        if a[mini] > a[j]:
            mini = j
    a[i], a[mini] = a[mini],a[i]
    
print(a)
```

# 삽입정렬

```python
a= [7,5,9,0,3,1,6,2,4,8]


for i in range(len(a)):
    for j in range(i,0,-1): ## 인덱스 i부터 1까지 1씩감소 하면서 반복
        if a[j] < a[j-1]:   ## 한칸씩 왼쪽으로 이동
            a[j],a[j-1] = a[j-1],a[j]
        else: # 자기보다 작은 데이터를 만나면 멈춤
            break
            
print(a)
```

# 퀵정렬

```python
# 1.특정 숫자를 가운대로 양옆이 작은수와 큰수로 그룹으로나뉨
# 2.왼쪽 데이터 묶음
# 3.오른쪽 데이터 묶음



array= [5,7,9,0,3,1,6,2,4,8]

def quick_sort(array,start,end): # 원소가 1개인 경우 종료
    if start>= end:
        return
    pivot = start # 피벗의 첫번째 원소 
    left = start+1
    right = end
    
    while(left<=right):
        #피벗보다 큰 데이터를 찾을때까지 반복
        while(left<=end and array[left]<= array[pivot]):
            left+=1
        
        #피벗보다 작은 데이터를 찾을때까지 반복
        while(right<=start and array[right]<= array[pivot]):
            right-=1
            
        if(left>right) : # 엇갈렸다면 작은 데이터와 피벗을 교체
            array[right], array[pivot] = array[pivot],array[right]
        else:            # 엇갈리지 않았다면 작은 데이터와 큰데이터 을 교체
            array[left], array[pivot] = array[pivot],array[left]
    
    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬수행
    quick_sort(array,start,right-1)
    quick_sort(array,right+1,end)
    

quick_sort(array,0,len(array)-1)

print(array)
```

# 계수정렬

가장 작은 데이터부터 가장 큰 데이터까지 범위가 있을때 리스트 생성하고
count를 세어가며 정렬하는 정렬법

(성적1~100점)까지

```python
array = [7,5,9,0,3,1,6,2,9,1,4,8,0,5,2]

count = [0]* (max(array)+1)

for i in range(len(array)):
    count[array[i]] +=1 # 각 데이터에 해당하는 인덱스값 증가
    
for i in range(len(count)): # 리스트에 기록된 정렬정보 확인
    for j in range(count[i]):
        print(i,end='') # 띄어쓰기를 구분으로 등장한 횟수만큼 인덱스출력
```
