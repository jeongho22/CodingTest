# Queue

```python

#  

from collections import deque

queue = deque()


queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)

print(queue)

#
queue.popleft()

#
queue.append(1)
#
queue.append(4)

#
queue.popleft()


print(queue)




#
queue.reverse()
print(queue)
```

```python

n = int(input())

#  

# 1. 
def good(n):
    
    if n<=1:
        return 1
    else:
        return n*good(n-1)
        
good(n)
```

```python
n = int(input())

#2. 

def good(n):
    
     
    if n<=1:
        print(1)
    
    else:
        result=1   
        for i in range(1,n+1):
            result*=i
        print(result)
            
            
good(n)
```
