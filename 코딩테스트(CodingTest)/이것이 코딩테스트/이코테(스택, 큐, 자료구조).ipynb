# Stack

```python
stack=[]

stack.append(5)
stack.append(2)
stack.append(3)
stack.append(1)


print(stack)

stack.pop()

print(stack)

stack.pop()

print(stack)


# 뒤에서부터
print(stack[::-1])
```

# Queue

```python

#들어온 순서대로 나감

from collections import deque

queue = deque()


queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)

print(queue)

#삭제
queue.popleft()

#추가
queue.append(1)
#추가
queue.append(4)

#삭제
queue.popleft()


print(queue)




#결론
queue.reverse()
print(queue)
```

```python

n = int(input())

# 팩토리얼 

# 1.재귀 호출
def good(n):
    
    if n<=1:
        return 1
    else:
        return n*good(n-1)
        
good(n)
```

```python
n = int(input())

#2. 반복호출

def good(n):
    
     
    if n<=1:
        print(1)
    
    else:
        result=1   
        for i in range(1,n+1):
            result*=i
        print(result)
            
            
good(n)
```
